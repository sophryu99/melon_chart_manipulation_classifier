# 데이터 전처리
1. 다른 가수 곡 정보 삭제 (곡 제목만 남기기)
2. 초성 치환
3. 이모티콘 삭제
4. 리뷰 길이 128자로 제한
5. 관련성 없는 리뷰 드랍

```python
import pandas as pd
import emoji
```

```python
#data = pd.read_csv("melon_akmu.csv", index_col = 0)
data = pd.read_csv("melon_limjaehyun.csv", index_col = 0)
data.head(10)
```

```python
def remove_emoji(text):
    return emoji.get_emoji_regexp().sub(u'', text)

info = ["재생 다운로드 곡명", "재생 다운로드 앨범명", "재생 >다운로드 앨범명", "YouTube에서", "(커버)", "뮤비해석", "[", "]", 'Feat. 은하 Of 여자친구', '♩', '♬', '♪', "(", ")"]
words = {"ㄴㄱ": "누구", "ㅇㅈ":"인정", "ㅆㄹㄱ": "쓰레기", "ㅅㅈㄱ": "사재기", "ㄹㅇ": "진짜", "ㅈㅂ": "제발", "ㄲㅈ": "꺼져", "ㅂㅌㅅㄴㄷ": "bts", "방탄": 'bts', "ㅂㅃㄱㅅㅊㄱ": "볼빨간 사춘기", "볼사": "볼빨간 사춘기"}

# 관련성 없는 리뷰 드랍
data = data[data.label != 2]

def process(df):
    
    # 타 가수 정보 제목 외에 제거
    for j in range(len(info)):
        df["comment"] = df["comment"].str.replace(info[j], "")
        df["comment"] = df["comment"].str.replace(info[j], "")
    df["comment"] = df["comment"].str.split("아티스트명").str[0]
        
    # 초성, 단어 치환
    for k, v in words.items():
        df["comment"] = df["comment"].str.replace(k, v)
    
    # 이모티콘 삭제
    sentences = df['comment']
    deemoji = [remove_emoji(s) for s in sentences]
    df["comment"] = deemoji
    
    # 리뷰 120자 제한   
    limit = [s[0:128:] for s in sentences]
    df['comment'] = limit
    
```

```python
process(data)
```

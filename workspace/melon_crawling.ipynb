```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import re

wd = "/Users/sehwaryu/Documents/DSL/chrome/chromedriver"    

driver = webdriver.Chrome(wd)
name_list=[]
artist_list=[]
comment_list=[]
comment__list=[]
rating_list=[]   #평점
jjim_list=[]     #찜 수
comment_time_list = []
comment__time__list = []


# 악뮤, 임재현
song_url=[10331947,10331499]

page = 2 # 한 페이지당 약 10개의 댓글이 표시
for song in song_url:
    for i in range(1,page+1):
                # 추천순
                addr = f"https://www.melon.com/album/detail.htm?albumId={song}#cmtpgn=&pageNo={i}&sortType=1&srchType=2&srchWord="
                driver.get(addr)
                time.sleep(1.5)

                html = driver.page_source
                dom = BeautifulSoup(html, "lxml")
                # 리뷰 
                comments_raw = dom.find_all("div", {"class" : "cmt_text d_cmtpgn_cmt_full_contents"})
                comments = [comment.text for comment in comments_raw]
                comment__list.append(comments)
                # 리뷰 작성 시간
                comments_date = dom.find_all("span", {"class" : "date"})
                comments_date = [date.text for date in comments_date]
                comment__time__list.append(comments_date)
                print(str(i)+"페이지", end=" ")

    for k in range(0,len(sum(comment__list,[]))):
        name = driver.find_elements_by_css_selector(
        #'#downloadfrm > div.section_info > div > div.entry > div.info > div.song_name'
        '#conts > div.section_info > div > div.entry > div.info > div.song_name'
        )[0].text
        name_list.append(name)
    
        artist = driver.find_elements_by_css_selector(
        '#conts > div.section_info > div > div.entry > div.info > div.artist > a > span:nth-child(1)'
        )[0].text
        artist_list.append(artist)
    
        rating = driver.find_elements_by_css_selector(
        '#gradPointLayer'
        )[0].text
        rating_list.append(rating)
        jjim = driver.find_elements_by_css_selector(
        '#d_like_count'
        )[0].text
        jjim_list.append(jjim)
    comment_list.append(sum(comment__list,[]))
    comment_time_list.append(sum(comment__time__list,[]))
    

comment_list=sum(comment_list,[])
comment_time_list = sum(comment_time_list, [])

for j in range(0, len(comment_list)):
    comment_list[j]=re.sub('\t+', ' ', comment_list[j])
    comment_list[j]=re.sub(' +', ' ', comment_list[j])
    comment_list[j]=re.sub('내용', '', comment_list[j])
```

```python
df = pd.DataFrame([x for x in zip(name_list,artist_list,comment_list,comment_time_list,jjim_list,rating_list)],columns=["song","artist","comment","comment_date", "jjim","rating"])
```
